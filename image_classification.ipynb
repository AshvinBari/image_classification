```python
import torch
from PIL import Image
from transformers import CLIPProcessor, CLIPModel

# CLIP model and processor
model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32")
processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")

# Load the image
def load_image(image_path):
    return Image.open(image_path)

# Define a list
possible_descriptions = ["a picture of medicine", "a picture of food", "a picture of an animal", "a picture of a person", "sea","cake"]

# Classify the image
def classify_image(image_path):
    image = load_image(image_path)

    # Prepare inputs for the model (image + text descriptions)
    inputs = processor(text=possible_descriptions, images=image, return_tensors="pt", padding=True)

    # Perform forward pass to get image-text similarity scores
    outputs = model(**inputs)
    logits_per_image = outputs.logits_per_image
    probs = logits_per_image.softmax(dim=1)

    # Get the best description
    predicted_index = probs.argmax().item()
    predicted_description = possible_descriptions[predicted_index]

    return predicted_description

# Ex
image_path = "/content/cake.webp"
prediction = classify_image(image_path)
print(f"Predicted description: {prediction}")
```
